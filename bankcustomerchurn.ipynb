```python
import pandas as pd
data=pd.read_csv("/kaggle/input/bank-customer-churn-dataset/Bank Customer Churn Prediction.csv")
data.sample()
```

```python
from sklearn.model_selection import train_test_split

X=data.drop(columns=["churn"])
y=data.churn
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3)
X_train.head()
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder,StandardScaler

categoricalPipeline=Pipeline([
    ("ohe",OneHotEncoder(drop="first",sparse_output=False))
])
numericalPipeLine=Pipeline([
    ("ss",StandardScaler())
])
categorical=["country","gender","products_number","credit_card","active_member"]
numerical=["credit_score","age","tenure","balance","estimated_salary"]
transformer=ColumnTransformer([
    ("cat",categoricalPipeline,categorical),
    ("num",numericalPipeLine,numerical)
])
pipeline=Pipeline([
    ("preprocess",transformer)
])

X_train=pipeline.fit_transform(X_train)
X_test=pipeline.transform(X_test)

X_train=pd.DataFrame(X_train)
X_test=pd.DataFrame(X_test)
```

```python
X_train.head()
```

```python
import tensorflow
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense,Input,Dropout

model=Sequential()
model.add(Input(shape=(13,)))
model.add(Dense(40,activation="relu"))
model.add(Dense(20,activation="relu"))
model.add(Dense(10,activation="relu"))
model.add(Dense(1,activation="sigmoid"))

model.compile(optimizer="Adam",loss="binary_crossentropy",metrics=["accuracy"])
```

```python
model.fit(X_train,y_train,epochs=100)
```

```python
pred=model.predict(X_test)
pred=pd.DataFrame(pred,columns=["predChurn"])

result=pd.DataFrame({
    "Actual Value":y_test.values,
    "Predicted Value":pred.values.ravel(),
    "Predicted Class":(pred.values.ravel()>0.5).astype(int)
})
result.sample(10)
```
